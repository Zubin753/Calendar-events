## Часть 2. Вычисления в Python

Напишите функцию на Python, которая вычисляет математическую функцию $y(x)$ с заданными скалярными параметрами $a$, $b$, $c$, $d$ (если применимо) и списком значений независимой переменной $x$. Найти математические функции, доступные в библиотеке `numpy`, можно [здесь](https://numpy.org/doc/stable/reference/routines.math.html).

Пример для функции $y(x) = a \sin^2x + b\log_cx$ может выглядеть так:
```python
def my_function(x,a,b,c,d):
    return a * np.sin(x) ** 2 + b * np.log(x) / np.log(c)
```

````python
def my_function(x,a,b,c):
  ### BEGIN YOUR CODE
  return a * (np.sin(b*x)/x) + c
  ### END YOUR CODE
````

Задайте значения для параметров $a$, $b$, $c$, $d$ и укажите диапазон для переменной $x$:

````python
### BEGIN YOUR CODE

a1 = 2
b1 = 1
c1 = 3
a2 = 1
b2 = -1
c2 = 1
x = np.linspace(1, 10, 200)
print(x)

### END YOUR CODE
````

Вычислите значения функции $y(x)$ на интервале `x`, векторно и без использования циклов. Постройте график функции. Измените значения параметров $a$, $b$, $c$, $d$ и $x$ в ячейке выше, чтобы получить аккуратный, понятный и красивый график.

````python
### BEGIN YOUR CODE


fig, axs = plt.subplots(1, 2, figsize=(10, 8))
y1 = my_function(x, a1, b1, c1)
axs[0].plot(x, y1, color='red', linestyle='--', label=f'{a1} * (np.sin({b1}*x)/x) + {c1}')
axs[0].set_title(f'График 1: {a1} * (np.sin({b1}*x)/x) + {c1}')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].legend()
axs[0].grid()

for value in range(1, 11):  # Отметка от 1 до 20
        y_value = my_function(value, a1, b1, c1)  # Находим y для данного x
        axs[0].scatter(value, y_value, color='black', s=20, zorder=5)
        axs[0].text(value+1, y_value + 0.02, f'({value}, {y_value:.2f})', fontsize=8, ha='center')
y2 = my_function(x, a2, b2, c2)
axs[1].plot(x, y2, color='blue', linestyle=':', label=f'{a2} * (np.sin({b2}*x)/x) + {c2}')
axs[1].set_title(f'График 2: {a2} * (np.sin({b2}*x)/x) + {c2}')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].legend()
axs[1].grid()

for value in range(1, 11):  # Отметка от 1 до 20
        y_value = my_function(value, a2, b2, c2)  # Находим y для данного x
        axs[1].scatter(value, y_value, color='black', s=20, zorder=5)
        axs[1].text(value+1, y_value + 0.02, f'({value}, {y_value:.2f})', fontsize=8, ha='center')

plt.tight_layout()
plt.show()

### END YOUR CODE
````
